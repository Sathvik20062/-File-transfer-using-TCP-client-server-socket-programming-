```python
import matplotlib.pyplot as plt
import sympy,numpy as np
import math
```

```python
class convexHull:
    def __init__(self,points):
        self.pointSet=points
        self.Xorg_cood = [i[0] for i in self.pointSet]
        self.Yorg_cood = [i[1] for i in self.pointSet]
        self.x=sympy.Symbol('x')
        self.y=sympy.Symbol('y')
        self.x1=sympy.Symbol('x1')
        self.y1=sympy.Symbol('y1')
        self.x2=sympy.Symbol('x2')
        self.y2=sympy.Symbol('y2')
        self.eqn=(self.y2-self.y1)*self.x+(self.x1-self.x2)*self.y-(self.x1*self.y2-self.x2*self.y1)
        self.ExtremeEdges = []
        self.Extremepoints = []
        self.validity=[]
        self.findExtremes()
         
        self.ShowPolygon()
    
    def findExtremes(self):
       
       for i in range(len(self.pointSet)):
           eqn=self.eqn
           # Substituting x2,y2 
           eqn=eqn.subs([(self.x2,self.pointSet[i][0]),(self.y2,self.pointSet[i][1])])
           
           for j in range(i+1,len(self.pointSet)):
               
               # Check if it is already an Extream
                if [self.pointSet[j]] in self.validity:
                    if self.pointSet[j] not in self.Extremepoints:
                      self.Extremepoints.append(self.pointSet[j])
                    continue
                # Check if it is already an Extream 
                # Substituting x1,y1
                eqn1=eqn.subs([(self.x1,self.pointSet[j][0]),(self.y1,self.pointSet[j][1])])
                value=[]
                valid=[]
                for k in range(len(self.pointSet)):
                    if k!=i and k!=j:
                      # Substituting x,y 
                      eqn2=eqn1.subs([(self.x,self.pointSet[k][0]),(self.y,self.pointSet[k][1])])
                      value.append(eqn2)
                      
                      # Multiple points on same line 
                      if eqn2 == 0:
                         valid+= [self.pointSet[k]]
                      # Multiple points on same line   
             #  Checking for the validity of line segment
             #   1 => point above line segment
             #   0 => point on line segment
             #  -1 => point below line segment
                          
                check=[1 if i>0 else -1 if i<0 else 0 for i in value] 
         # <--- Checking for the validity of line segment
         # if all points other than line segment are 
         # 1. above and on the line segment and not below  1 and 0 and not -1
         # 2. only above the line segment and not below    1 and not -1
         # 3. below and on the line segment and not above -1 and 0 and not 1
         # 4. only below the line segment and not above   -1 and not 1
         # Then the that line segment is an valid extream side of covex hull
         
                if (((1 in check and 0 in check) and (-1 not in check)) or ((1 in check) and (-1 not in check) )) or (((-1 in check and 0 in check) and (1 not in check)) or ((-1 in check) and (1 not in check) )) :
                    self.ExtremeEdges.append([self.pointSet[i],self.pointSet[j]])
                    
                    # <--- Valide Multiple points on same Extreame line 
                    self.validity.append(valid)
                    # Multiple points on same Extreame line 
                    
                    # <--- Keeping Track of Extream points
                    if self.pointSet[i] not in self.Extremepoints:
                       self.Extremepoints.append(self.pointSet[i])
                    if self.pointSet[j] not in self.Extremepoints:
                       self.Extremepoints.append(self.pointSet[j])
                    # Keeping Track of Extream points
         #  Checking for the validity of line segment
       print(self.Extremepoints)
  
    
    def ShowPolygon(self):
        axes1=plt.axes()
        plt.scatter(self.Xorg_cood, self.Yorg_cood,axes=axes1)
        # <--- for all pair of points in Extremes plotting line ---->
        for coordinate in self.ExtremeEdges:
            xExtreme = [pointX[0] for pointX in coordinate]
            yExtreme = [pointX[1] for pointX in coordinate]
            plt.plot(xExtreme, yExtreme,axes=axes1)
        # <--- for all points in Extremes printing tags ---->
        for pointIndex in range(len(self.Extremepoints)):
            plt.text(self.Extremepoints[pointIndex][0]+0.3,self.Extremepoints[pointIndex][1]+0.4, '({a},{b})'.format(a=self.Extremepoints[pointIndex][0], b=self.Extremepoints[pointIndex][1]), fontsize = 12)
        plt.xlim([0, 20])
        plt.ylim([0, 10])
        
points=[(10,0),(11,5),(5,3),(9,3.5),(15,3),(12.5,7),(6,6.5),(7.5,4.5),(13.75,5),(5.5,4.75)]
poly=convexHull(points)
```

```python
class ConvexHullDC:
    def __init__(self,points):
        self.pointSet=points
        self.Xorg_cood = [i[0] for i in self.pointSet]
        self.Yorg_cood = [i[1] for i in self.pointSet]
        pointsort = sorted(self.pointSet,key=lambda x: x[0])
        S1,S2=self.Divide(pointsort)
        
        
    def Divide(self,pointsort):
        if len(pointsort)<=5:
            return pointsort
        mid = int(len(pointsort)/2)
        S1 = self.Divide(pointsort[:mid])
        S2 = self.Divide(pointsort[mid:])
        print(f"Set 1 {S1}")
        print(f"Set 2 {S2}")
        C1 = self.findHull(S1)
        C2 = self.findHull(S2)
        upperTang = self.findtangent(C1,C2,"upper")
        lowerTang = self.findtangent(C1,C2,"lower")
        self.Merging(C1,C2,upperTang,lowerTang)
        return S1,S2

    def Merging(self,C1,C2,up,low):
        
        Etremes=[x for x in C1 if x[0] <= up[0][0] and x[0] <= low[0][0]]+[up[0],up[1]]
        Etremes+=[x for x in C2 if x[0] >= up[1][0] and x[0] >= low[1][0]]+[low[1],low[0]]
        Etremes = list(set(Etremes))
        
        # <--------------- Sort Points list to plot polygon without intersection in matplotlib
        # compute centroid
        cent=(sum([p[0] for p in Etremes])/len(Etremes),sum([p[1] for p in Etremes])/len(Etremes))
       # sort by polar angle
        Etremes.sort(key=lambda p: math.atan2(p[1]-cent[1],p[0]-cent[0])) 
        # ----------------------->
        
        C1 = sorted(C1,key=lambda x: x[0])
        C2 = sorted(C2,key=lambda x: x[0])
        # <-- ploting C1,C2 hull with tangent-->
        plt.figure()
        axes1=plt.axes()
        for hull in [C1,C2]:
            
            x = [pointX[0] for pointX in hull]
            y = [pointX[1] for pointX in hull]
            plt.scatter(x,y,axes=axes1)
            plt.plot(x+[x[0]],y+[y[0]],axes=axes1)
        for tangent in [up,low]:
            xtan = [pointX[0] for pointX in tangent]
            ytan = [pointX[1] for pointX in tangent]
            plt.plot(xtan,ytan,axes=axes1)
            
        # <-- ploting convexhull -->
        plt.figure()
        axes2=plt.axes()
        x = [pointX[0] for pointX in Etremes]
        y = [pointX[1] for pointX in Etremes]
        plt.scatter(x,y,axes=axes2)
        plt.plot(x+[x[0]],y+[y[0]],axes=axes2)
           
        
    def findHull(self,points):
        #convexhull findextremes() [LAB-8]
        hull=convexHull(points)
        hull.findExtremes()
        plt.figure()
        hull.ShowPolygon()
        return hull.Extremepoints
    
    def findtangent(self,C1,C2,pose="upper"):
        C1sort=sorted(C1,key=lambda x: x[1])
        C2sort=sorted(C2,key=lambda x: x[1])
        if pose =="upper":
          p1=C1sort[len(C1sort)-1]
          p2=C2sort[len(C2sort)-1]
        elif pose =="lower":
          p1=C1sort[0]
          p2=C2sort[0]
        else:
          print("Invalid pose")
        return (p1,p2)

points=[(10,0),(11,5),(5,3),(9,3.5),(15,3),(12.5,7),(6,6.5),(7.5,4.5)]
shape=ConvexHullDC(points)
```
